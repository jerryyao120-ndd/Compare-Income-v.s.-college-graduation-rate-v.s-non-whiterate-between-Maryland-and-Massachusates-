```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("C:\\Users\\jerryyao\\Downloads\\forpy.csv")
ma=data[(data['State']==' MA')]
md=data[(data['State']==' MD')]
plt.scatter(ma['College_Graduation_Rate_rP_gP_pall'],ma['Household_Income_rP_gP_pall'])
plt.scatter(md['College_Graduation_Rate_rP_gP_pall'],md['Household_Income_rP_gP_pall'])
plt.xlabel('College_Graduation_Rate')
plt.ylabel('Household_Income')
plt.legend(['MA','MD'])
plt.figure()
plt.scatter(ma['Fraction_Non-White_in_2010'],ma['Household_Income_rP_gP_pall'])
plt.scatter(md['Fraction_Non-White_in_2010'],md['Household_Income_rP_gP_pall'])
plt.xlabel('Fraction_Non-White')
plt.ylabel('Household_Income')
plt.legend(['MA','MD'])
```
